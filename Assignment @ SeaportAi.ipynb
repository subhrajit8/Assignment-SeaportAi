### Combined the 3 Datasets

```python
data = pd.concat([oct_data,nov_data,dec_data],axis=0)
```

```python
data
```

```python
data.reset_index(inplace=True)
data
```

```python
## Dropped unwanted attributes.
data.drop(['index','Unnamed: 0'],axis=1,inplace=True)
data
```

```python
data.isnull().sum()
```

```python
## Dropped records with majority of attributes having NULL values
data.dropna(subset=['event_type','product_id','category_id','price','user_id'],inplace=True)
```

```python
data.isnull().sum()
```

```python
data.nunique()
```

```python
data.event_type.unique()
```

```python
## Dropped 'Category_code' attribute as majority of values of this attribute contains NULL values.
data.drop('category_code',axis=1,inplace=True)
data
```

```python
data.isnull().sum()
```

###### Most Expensive Product

```python
exp_pro=data[['product_id','brand','price']][data.price == data.price.max()]
exp_pro.drop_duplicates(inplace=True)
exp_pro
```

```python
print("The most expensive product : " + "Product ID : " + str(exp_pro.iloc[0,0]) + " Price : " + str(exp_pro.iloc[0,2]) + " Brand : " + str(exp_pro.iloc[0,1]))
```

###### Most Expensive Brands

```python
a=data.groupby(['brand'])['price'].mean()
a.sort_values(ascending=False,inplace=True)
a=a.reset_index()
a
```

```python
print("Most Expensive Brand : " + str(a.iloc[0,0]))
```

###### Most Economical Brands

```python
b=data.groupby(['brand'])['price'].mean()
b.sort_values(ascending=True,inplace=True)
b=b.reset_index()
b
```

```python
print("Most Economical Brand : " + str(b.iloc[0,0]))
```

### Purchased Products
We parsed the data from the main datasets for the products which were purchased by the Users.

```python
purchase=data.loc[data['event_type']=='purchase']
purchase
```

###### Top Purchased Brands

```python
brandpur=purchase['brand'].value_counts().reset_index()
brandpur.columns=['Brand','Item Purchased']
brandpur.head(10)
```

```python
brandpur.head(10).plot('Brand','Item Purchased',kind='bar',figsize=(13,7))
plt.ylabel('Item Purchased')
plt.title('Top 10 Purchased brand.')
```

```python
print("Most Favourite Brand : " + str(brandpur.iloc[0,0]))
```

###### Top Buyers

```python
userpur=purchase['user_id'].value_counts().reset_index()
userpur.columns=['User ID','Items Purchased']
userpur.head(10)
```

```python
userpur.head(10).plot('User ID','Items Purchased',kind='bar',figsize=(13,7))
plt.ylabel('Items Purchased')
plt.xlabel('User ID')
plt.title('Top 10 viewed product.')
```

```python
print("Top User : User ID = " + str(userpur.iloc[0,0]))
```

###### Most Purchased Products

```python
propur=purchase['product_id'].value_counts().reset_index()
propur.columns=['Product ID','Total Sell']
propur.head(10)
```

```python
print("Most purchased product : Product ID = " + str(propur.iloc[0,0]))
```

###### Top expensive products purchased

```python
exppur=purchase[['product_id','brand','price']].sort_values(by=['price'],ascending=False)
exppur.drop_duplicates(inplace=True)
exppur.head(10)
```

```python
print("Most expensive product purchased: Product ID = " + str(exppur.iloc[0,0]) + " with Price = " + str(exppur.iloc[0,2]) + " of " + str(exppur.iloc[0,1]) + " brand.")
```

###### Top economical products purchased

```python
ecopur=purchase[['product_id','brand','price']].sort_values(by=['price'])
ecopur.drop_duplicates(inplace=True)
ecopur.head(10)
```

```python
print("Most economical product purchased: Product ID = " + str(ecopur.iloc[0,0]) + " with Price = " + str(ecopur.iloc[0,2]) + " of " + str(ecopur.iloc[0,1]) + " brand and " + " Product ID = " + str(ecopur.iloc[1,0]) + " with Price = " + str(ecopur.iloc[1,2]) + " of " + str(ecopur.iloc[1,1]) + " brand.")
```

###### User who spent the most|

```python
userspent=purchase.groupby('user_id')['price'].sum().reset_index()
userspent.columns=['User ID','Amount Spent']
userspent.sort_values(by='Amount Spent',ascending=False,inplace=True)
userspent.head(10)
```

```python
userspent.head(10).plot('User ID','Amount Spent',kind='bar',figsize=(13,7))
plt.ylabel('Amount Spent')
plt.xlabel('User ID')
plt.title('Top 10 higher spender.')
```

```python
print("Highest Spender : User ID = " + str(userspent.iloc[0,0]))
```
